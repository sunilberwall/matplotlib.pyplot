## matplotlib

```python
import matplotlib.pyplot as plt
```

```python
# plot graph
x=[1,2,3,4,5,6]
y=[21,22,23,24,24,25]
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("graph")
plt.show()
```

```python
# get a grid
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("graph")
plt.grid()
plt.show()
```

```python
# get a legend
plt.plot(x,y,label="line")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("graph")
plt.legend()
plt.show()
```

```python
# change linestyle
plt.plot(x,y,linestyle="--")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("graph")
plt.show()
```

```python
# color change
plt.plot(x,y,color="red")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("graph")
plt.show()
```

```python
# add a marker
plt.plot(x,y,marker="<")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("graph")
plt.show()
```

```python
# set linewidth
plt.plot(x,y,linewidth=20)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("graph")
plt.show()
```

```python
# all style present in mat.
plt.style.available
```

```python
# use style
plt.style.use("fivethirtyeight")
plt.plot(x,y,linewidth=20)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("graph")
plt.show()
```

```python
# bar plot
plt.bar(x,y)
plt.show()
```

```python
import numpy as np
x=[51,55,58,67,69,72]
y=[23,22,23,24,28,23]
y1=[36,37,38,39,10,11]
y2=[9,8,7,6,5,4]
x_index=np.arange(len(x))
width=0.2
plt.bar(x_index-width/2,y,label="y")
plt.bar(x_index,y1,label="y1")
plt.bar(x_index+width,y2,label="y2")
# here put x's value on x-axis
plt.xticks(ticks=x_index,labels=x)
plt.legend()
plt.show()
```

```python
# stack plots
x=[51,55,58,67,69,72]
y=[23,22,23,24,28,23]
y1=[36,37,38,39,10,11]
y2=[9,8,7,6,5,4]
plt.stackplot(x,y,y1,y2)
plt.show()
```

```python
# stack plots
x=[51,55,58,67,69,72]
y=[23,22,23,24,28,23]
y1=[36,37,38,39,10,11]
y2=[9,8,7,6,5,4]
l=["y","y1","y2"]
plt.stackplot(x,y,y1,y2,labels=l)
plt.legend()
plt.show()
```

```python
# stack plots
x=[51,55,58,67,69,72]
y=[23,22,23,24,28,23]
y1=[36,37,38,39,10,11]
y2=[9,8,7,6,5,4]
plt.stackplot(x,y,y1,y2,colors=["red","blue","green"])
plt.show()
```

```python
# filling area on line plot
plt.plot(x,y)
plt.fill_between(x,y,alpha=0.5)
plt.show()
```

```python
plt.plot(x,y)
plt.fill_between(x,y,alpha=0.5,where=2>15,interpolate=True)
plt.show()
```

```python
# histogram
plt.hist(x)
plt.show()
```

```python
plt.hist(x,bins=4,edgecolor="green")
```

```python
# scatter plot
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,s=100,c="green",marker="x",cmap="Greens")
plt.show()
```

```python
plt.scatter(x,y,s=100,c="green",marker="x",cmap="Greens")
cbar=plt.colorbar()
cbar.set_label("data")
plt.show()
```
